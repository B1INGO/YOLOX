# 预处理原始数据集合

```python
import os
import random
```

```python
images_path = "JPEGImages/"
xmls_path = "Annotations/"
train_val_txt_path = "ImageSets/Main/"
val_percent = 0.1
```

```python
images_list = os.listdir(images_path)
for i in range(3): random.shuffle(images_list)
```

```python
# 划分验证集和训练集
val_images_count = int(val_percent*len(images_list))
train_images_count = len(images_list) - val_images_count
```

```python
# 划分训练集-->train.txt
with open(os.path.join(train_val_txt_path,"train.txt"),"w") as train_txt:
    train_count = 0
    for i in range(train_images_count):
        text = images_list[i].split(".jpg")[0] + "\n"
        train_txt.write(text)
        train_count += 1
        print("train_count:" + str(train_count))
```

```python
# 划分验证集-->val.txt
with open(os.path.join(train_val_txt_path,"val.txt"),"w") as val_txt:
    val_count = 0
    for i in range(val_images_count):
        text = images_list[train_images_count + i].split(".jpg")[0] + "\n"
        val_txt.write(text)
        val_count += 1
        print("val_count:",val_count)
```
